```python
import pandas as pd 
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
df0 = pd.read_excel('/Volumes/lab-windingm/home/users/cochral/plug_camera/staging/staging-paramaters/2024_05_30_4h_staging_6020_5020.xlsx')

df1 = pd.read_excel('/Volumes/lab-windingm/home/users/cochral/plug_camera/staging/staging-paramaters/2024_06_07_4h_staging_6020_5020.xlsx')
```

```python
df = pd.concat([df0, df1], ignore_index=True) 
```

```python
df['condition'] = df['condition'].str.replace(r'\s+', '', regex=True)
```

```python


df
```

```python
plt.figure(figsize=(6,4))

sns.barplot(data=df, x='age_of_flies_on_day_of_staging', y='pupae_count', hue='condition', edgecolor='black', ci='sd', palette='Blues_d')

plt.ylabel('Pupae', fontweight='bold')
plt.xlabel('Age of Flies', fontweight='bold')
plt.title('Pupae Comparison for 6020 and 5020 4h Staging', fontweight='bold')

plt.legend(title='Condition', bbox_to_anchor=(1, 1), loc='upper left')

plt.show()
```

```python
df['date_of_virgin_collection'] = df['date_of_virgin_collection'].astype(str)

# Replace '2024-05-21' with 'Tue' in the 'date_of_virgin_collection' column
df['date_of_virgin_collection'] = df['date_of_virgin_collection'].replace('2024-05-21', 'Tue')

# Replace '2024-05-21' with 'Tue' in the 'date_of_virgin_collection' column
df['date_of_virgin_collection'] = df['date_of_virgin_collection'].replace('2024-05-22', 'Wed')


df['date_of_virgin_collection'] = df['date_of_virgin_collection'].replace('2024-05-29', 'Wed')

df.rename(columns={'date_of_virgin_collection': 'collection day'}, inplace=True)

df
```

```python
plt.figure(figsize=(3, 6))

sns.scatterplot(data=df, x='age_of_flies_on_day_of_staging', y='pupae_count', hue='condition', edgecolor='black', palette='magma', style='collection day', alpha=0.5, s=100)

# x axis only 9 and 10
plt.xticks([9, 10])

plt.ylabel('Pupae', fontweight='bold')
plt.xlabel('Age of Flies', fontweight='bold')
plt.title('Pupae Comparison for 6020 and 5020 4h Staging', fontweight='bold')


plt.legend(title='Condition and Collection day', bbox_to_anchor=(1, 1), loc='upper left')



plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))

sns.scatterplot(data=df, y='manual day 4 count', x='pupae_count', hue='age_of_flies_on_day_of_staging', palette='Blues_d', style='collection day', s=80)

plt.title('Influence of Pupae Number on Hole Formation ', fontweight='bold')
plt.xlabel('Total Pupae', fontweight='bold')
plt.ylabel('Number of Holes on Day 4', fontweight='bold')


plt.legend(title='Condition and Collection day', loc='upper left', bbox_to_anchor=(1, 1))

plt.tight_layout()

plt.show()
```
