```python
import pandas as pd 
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
df0 = pd.read_excel('/Volumes/lab-windingm/home/users/cochral/plug_camera/staging/staging-paramaters/2024_03_26_4h_staging_6020.xlsx')

df1 = pd.read_excel('/Volumes/lab-windingm/home/users/cochral/plug_camera/staging/staging-paramaters/2024_04_17_4h_staging_6020.xlsx')

df2 = pd.read_excel('/Volumes/lab-windingm/home/users/cochral/plug_camera/staging/staging-paramaters/2024_04_24_4h_staging_6020.xlsx')

df3 = pd.read_excel('/Volumes/lab-windingm/home/users/cochral/plug_camera/staging/staging-paramaters/2024_05_01_4h_staging_6020.xlsx')
```

```python
df = pd.concat([df0, df1, df2, df3], ignore_index=True)
```

```python
sns.barplot(data=df, x='age_of_flies_on_day_of_staging', y='pupae_count', edgecolor='black', ci='sd', palette='Blues_d')

plt.title('Pupae Count for Vials After 4-Hour Staging at a 60:20 Ratio', fontweight='bold')
plt.xlabel('Age of Females', fontweight='bold')
plt.ylabel('Mean Total Pupae', fontweight='bold')

plt.tight_layout()
plt.savefig('/Users/cochral/repos/behavioural-analysis/plots/plug-camera/staging/4h 6020/Barchart_for_pupa.png', dpi=300, format='png')
 
plt.show()
```

```python
sns.scatterplot(data=df, x='net loss for current staging', y='pupae_count', hue='age_of_flies_on_day_of_staging', palette='viridis', edgecolor='darkgrey', alpha=0.9, s=50)

plt.title('The Effect of Loss on Pupae Production', fontweight='bold')
plt.ylabel('Pupae Count', fontweight='bold')
plt.xlabel('Net Loss at Staging', fontweight='bold')


plt.legend(title='Age of Flies', bbox_to_anchor=(1, 0.9))

plt.savefig('/Users/cochral/repos/behavioural-analysis/plots/plug-camera/staging/4h 6020/Netloss_pupa.png', dpi=300, format='png')

plt.show()
```

```python
plt.figure(figsize=(6, 4))

sns.scatterplot(data=df, y='manual day 4 count', x='pupae_count', hue='age_of_flies_on_day_of_staging', palette='viridis')

plt.title('Influence of Pupae Number on Hole Formation ', fontweight='bold')
plt.xlabel('Total Pupae', fontweight='bold')
plt.ylabel('Number of Holes on Day 4', fontweight='bold')


plt.legend(title='Age of Flies', loc='upper left', bbox_to_anchor=(1, 1))

plt.tight_layout()
plt.savefig('/Users/cochral/repos/behavioural-analysis/plots/plug-camera/staging/4h 6020/Pupa_hole_number.png', dpi=300, format='png')
plt.show()
```
