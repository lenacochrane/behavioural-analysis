PLOT GRAPHS FOR 18^C V 25^C DATA

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

18 V 25 VIRGIN

```python
virgin18 = [7, 27, 27, 11, 14, 2, 16, 11]

virgin25 = [15, 12, 9, 10, 10, 9, 11, 4]

times = ["23/01 9AM", "23/01 3.30PM", "24/01 9AM", "24/01 3.30PM", "25/01 9AM", "25/01 3.30PM", "26/01 9AM", "26/01 3.30PM"]

virgin = list(zip(times, virgin18, virgin25))

df_virgin = pd.DataFrame(virgin, columns=['times', 'virgin18', 'virgin25'])

df_virgin
```

```python
# want to plow 18 virgin v time and 25 virgin v time 

plt.plot(df_virgin["times"], df_virgin["virgin18"], color='blue' , label="Virgins Collected at 18")
plt.plot(df_virgin["times"], df_virgin["virgin25"], color='cyan' , label="Virgins Collected at 25")

plt.xlabel('time', fontweight='bold')
plt.ylabel("Number of Virgins Collected", fontweight='bold')
plt.title("Virgin Collection at 18 and 25", fontweight='bold')
plt.legend(loc=(0.8, 1))
plt.grid(True, color='gray', linestyle='--', linewidth=0.5)

plt.xticks(rotation=45)

plt.show()
```

18 V 25 TOTAL

```python
total = {'Time': ["23/01 9AM", "23/01 3.30PM", "24/01 9AM", "24/01 3.30PM", "25/01 9AM", "25/01 3.30PM", "26/01 9AM", "26/01 3.30PM"], '18': [34,	27,	27,	35,	35,	5,	35,	17], '25': [15,	63,	31,	44,	27,	28,	27,	25]}

df_total = pd.DataFrame.from_dict(total)

df_total
```

```python
# plot 

plt.plot(df_total['Time'], df_total['18'], color='lightgreen', label="Total Collected at 18")
plt.plot(df_total['Time'], df_total['25'], color='lightpink', label="Total Collected at 25")

plt.xlabel('time', fontweight='bold')
plt.ylabel("Total Collected", fontweight='bold')
plt.title("Total Collection at 18 and 25", fontweight='bold')
plt.legend(loc=(0.8, 1))
plt.grid(True, color='gray', linestyle='--', linewidth=0.5)

plt.legend(fontsize='small', shadow=True, fancybox=True, framealpha=0.9, borderpad=1)

plt.xticks(rotation=45)

plt.show()
```

18 V 25 RATIO VIRGIN/FEMALE 

```python

virgin_18 = [7, 27, 27, 11, 14, 2, 16, 11]

virgin_18_np = np.array(virgin_18)

total_18 = [34,	27,	27,	35,	35,	5,	35,	17]

total_18_np = np.array(total_18)

virgin_25 = [15, 12, 9, 10, 10, 9, 11, 4]

virgin_25_np = np.array(virgin_25)

total_25 = [15,	63,	31,	44,	27,	28,	27,	25]

total_25_np = np.array(total_25)

# want to divide virgin / total and create two new lists / think i need arrays for this 

ratio_18 = virgin_18_np / total_18_np

ratio_25 = virgin_25_np / total_25_np

# plot both of these against time 

times = ["23/01 9AM", "23/01 3.30PM", "24/01 9AM", "24/01 3.30PM", "25/01 9AM", "25/01 3.30PM", "26/01 9AM", "26/01 3.30PM"]

# plot straight away 

plt.plot(times, ratio_18, label='18', color='green')
plt.plot(times, ratio_25, label='25', color='pink')

plt.title('Virgin Females / Total Collected', fontweight='bold')
plt.xlabel('Time')
plt.ylabel('Ratio of Virgin Females to Total Flies')

plt.legend(loc=(0.8, 1))
plt.grid(True, color='gray', linestyle='--', linewidth=0.5)

plt.legend(fontsize='small', shadow=True, fancybox=True, framealpha=0.9, borderpad=1)

plt.xticks(rotation=45)

plt.show()
```

AM V PM COLLECTION

```python

# want to compare am collection v pm collection 
# bar chart AM V PM 


virgin_18 = [7, 27, 27, 11, 14, 2, 16, 11]

total_18 = [34,	27,	27,	35,	35,	5,	35,	17]

virgin_25 = [15, 12, 9, 10, 10, 9, 11, 4]

total_25 = [15,	63,	31,	44,	27,	28,	27,	25]


# want to iterate through the lists and create two new lists AM, PM 

def two_new_list(list):

    list_AM = []
    list_PM = []

    for index, element in enumerate(list):

        index + 1

        if (index % 2) == 1:
            list_PM.append(element)
        
        else:
            list_AM.append(element)
        

    AM = sum(list_AM)
    PM = sum(list_PM)
    
    return f" AM list = {AM} the PM list = {PM}"



# virgin18 = 61, 51
# total18 = 131, 84
# virgin25 = 45, 35
# total25 = 100, 160
   
```

```python
collection_dictionary = {'virgin 18': [61, 51], 'total 18': [131, 84], 'virgin 25': [45, 35], 'total 25': [100, 160]}

collection_df = pd.DataFrame(collection_dictionary)

collection_df.index = ['AM', 'PM']

collection_df
```

```python

# for some reason super annoying to plot in this format; use transpose operation to invert dataframe 

inverted_collection_df = collection_df.T


inverted_collection_df
```

```python
ax = inverted_collection_df.plot(kind='bar', figsize=(5, 5), color=['lightblue', 'lightgreen'], edgecolor='black', linewidth=2)

# Customize the plot

ax.set_ylabel('Number Collected', fontsize=12, weight='bold')
ax.set_title('Total Collection AM and PM at 18 and  25', fontsize=14, weight='bold')

# Grid lines and ticks
ax.grid(True, which='major', linestyle='--', linewidth='0.5', color='gray', alpha=0.25)

ax.tick_params(axis='both', which='major', labelsize=10)

ax.tick_params(axis="x", labelrotation=0)

# Legend
ax.legend(title='Time of Day', fontsize='8', title_fontsize='9')

# Tight layout for better spacing
plt.tight_layout()

plt.show()

```
