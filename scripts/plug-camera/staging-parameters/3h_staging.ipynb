```python
import pandas as pd 
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt
import seaborn as sns
import os
from statsmodels.formula.api import ols
import statsmodels.api as sm
```

```python
df = pd.read_excel('/Users/cochral/Dropbox (The Francis Crick)/staging/holes_pupae_counting_staging.xlsx', '26.02 3h')
```

```python
plt.figure(figsize=(6, 4))

df['Ratio'] = df['number of females'].astype(str) + ':' + df['number of males'].astype(str)

df_filter = df[df['Total Pupa'] != 0]

plt.figure(figsize=(8, 5))

sns.barplot(data=df_filter, x='Age of Females', y='Total Pupa', hue='Ratio', edgecolor='black', ci='sd', palette='Greens')

plt.title('Mean Total Pupae for Different Female Ages staged for 3h', fontweight='bold')
plt.xlabel('Age of Females', fontweight='bold')
plt.ylabel('Mean Total Pupae', fontweight='bold')

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.tight_layout()
plt.savefig('/Volumes/lab-windingm/home/users/cochral/plug_camera/staging/analysis/3h_staging_figures/pupa_3h_ratio.png', dpi=300, format='png')
 
plt.show()
```

```python
# convert to catagorical 


df_filter['Net Loss'] = df_filter['Net Loss'].astype(int)


df_filter['staging time'] = df_filter['staging time'].astype('category')
df_filter['Age of Females'] = df_filter['Age of Females'].astype('category')

# Create a FacetGrid for subplots based on 'staging time' and 'ratios' conditions
# 'col' and 'row' parameters could be adjusted based on your specific conditions.
g = sns.FacetGrid(df_filter, col='Ratio', hue='Age of Females', col_wrap=5, sharex=False, sharey=False, palette='Greens')

# Map a scatter plot to the FacetGrid
g.map(sns.scatterplot, 'Net Loss', 'Total Pupa')

# Add a legend

g.add_legend(title='Age of Females', bbox_to_anchor=(1, 0.5), loc='center left')



# Set the axis labels
g.set_axis_labels('Net Loss', 'Total Pupae')

# Adjust the subplot titles if necessary
g.set_titles(col_template="{col_name}")
g.set_titles(row_template='{row_name}', fontweight='bold')

plt.tight_layout()

plt.savefig('/Volumes/lab-windingm/home/users/cochral/plug_camera/staging/analysis/3h_staging_figures/net_loss_pupa_3h.png', dpi=300, format='png')
# Show the plot
plt.show()
```
