```python
import pandas as pd 
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
df0 = pd.read_excel('/Volumes/lab-windingm/home/users/cochral/plug_camera/staging/staging-paramaters/2024_03_12_4h_staging_5020.xlsx')

df1 = pd.read_excel('/Volumes/lab-windingm/home/users/cochral/plug_camera/staging/staging-paramaters/2024_03_19_4h_staging_5020_2.xlsx')

df2 = pd.read_excel('/Volumes/lab-windingm/home/users/cochral/plug_camera/staging/staging-paramaters/2024_03_26_4h_staging_6020.xlsx')

df3 = pd.read_excel('/Volumes/lab-windingm/home/users/cochral/plug_camera/staging/staging-paramaters/2024_04_11_4h_staging_6020_yeasttest.xlsx')
```

```python

df = pd.concat([df0, df1, df2, df3])
```

```python
df = df.reset_index()
```

```python
df.index.duplicated()
```

```python
df['age_of_flies_on_day_of_staging'] = df['age_of_flies_on_day_of_staging'].astype(str)
```

```python

df['Ratio'] = df['condition'].apply(lambda x: '50:20' if x == '4h staging; 50:20' else '60:20')


df
```

```python
# at the end bar graph comparing 6020 5020 (hue=ratio)


sns.barplot(data=df, x='age_of_flies_on_day_of_staging', y='pupae_count', hue='Ratio', edgecolor='black', ci='sd', palette='Blues_d')


plt.title('Pupae Comparison for Vials Staged for 4 Hours at 60:20 and 50:20 Ratio', fontweight='bold')
plt.xlabel('Age of Females', fontweight='bold')
plt.ylabel('Mean Total Pupae', fontweight='bold')

plt.tight_layout()
plt.savefig('/Users/cochral/repos/behavioural-analysis/plots/plug-camera/staging/4h 6020/Barchart_6020_5020_4h.png', dpi=300, format='png')
 
plt.show()

```

```python
# also plot for hole v pupa but no hue 

plt.figure(figsize=(6, 4))

sns.scatterplot(data=df, y='manual day 4 count', x='pupae_count', hue='Ratio', palette='viridis')

plt.title('Influence of Pupae Number on Hole Formation ', fontweight='bold')
plt.xlabel('Total Pupae', fontweight='bold')
plt.ylabel('Number of Holes on Day 4', fontweight='bold')


plt.legend(title='Age of Flies', loc='upper left', bbox_to_anchor=(1, 1))

plt.tight_layout()
plt.savefig('/Users/cochral/repos/behavioural-analysis/plots/plug-camera/staging/4h 6020/4h(6020_5020)_pupa_v_hole.png', dpi=300, format='png')
plt.show()
```
