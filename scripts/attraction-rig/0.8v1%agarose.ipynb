```python
plain_df = df[df['plain or side hole'] == 'plain']
```

```python
sample_plain = plain_df.groupby(['food or agarose', '0.8 or 1 % agaorse']).size().reset_index(name='count')

sample_plain
```

```python
plt.figure(figsize=(5, 4))

barplot = sns.barplot(data=plain_df, x='food or agarose', y='number underneath', hue='0.8 or 1 % agaorse', palette='Greens_d', edgecolor='grey')


plt.title('Number of Larvae Digging Underneath in Plain Petri', fontweight='bold')

plt.xlabel('Food or Agarose', fontweight='bold')
plt.ylabel('Number Underneath', fontweight='bold')

plt.legend(title='Agarose Concentration', loc='upper left')

annotations = ['n=6', 'n=6', 'n=12', 'n=9']


for i, bar in enumerate(barplot.patches):
    if i < len(annotations):  # Ensure we don't go out of range
        annotation = annotations[i]
        barplot.annotate(
            annotation,  # The custom annotation text
            (bar.get_x() + bar.get_width() / 2, bar.get_height()),  # Position the text at the top of the bar
            ha='center',  # Center the text horizontally
            va='center',  # Center the text vertically
            size=8,
            fontweight='bold',  # Set the font size to 8
            xytext=(0, 8),  # Offset the text by 8 points in the y-direction
            textcoords='offset points'  # Specify the offset in points
        )


plt.show()

```
