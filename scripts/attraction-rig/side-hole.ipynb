```python


# Directory containing the Excel files
directory = '/Volumes/lab-windingm/home/users/cochral/AttractionRig/data/pilot-hole-analysis/side-hole/agarose-n10'

# List all Excel files in the directory
excel_files = [f for f in os.listdir(directory) if f.endswith('.xlsx')]

# Number of Excel files
num_files = len(excel_files)

# Create subplots - we'll assume one row and multiple columns for simplicity
fig, axes = plt.subplots(nrows=1, ncols=num_files, figsize=(5*num_files, 5), squeeze=False)

# Loop over the Excel files and plot each on a separate subplot
for i, file in enumerate(excel_files):
    # Load the data
    file_path = os.path.join(directory, file)
    df = pd.read_excel(file_path)
    
    # Plot 'Number of Holes' vs 'Time'
    if 'time (s)' in df.columns and 'number in man made hole ' in df.columns:
        axes[0, i].plot(df['time (s)'], df['number in man made hole '])
        axes[0, i].set_xlabel('Time (S)')
        axes[0, i].set_ylabel('Number of Larvae in Hole')
        axes[0, i].set_title(file)
        axes[0, i].set_ylim(0, 10)
        axes[0, i].legend()

fig.suptitle('Agarose Side Hole N10', fontsize=20, fontweight='bold')


# Adjust layout
plt.tight_layout()



# Show the plot
plt.show()

```

FOOD SIDE HOLE

```python

# Directory containing the Excel files
directory = '/Volumes/lab-windingm/home/users/cochral/AttractionRig/data/pilot-hole-analysis/side-hole/food-n10'

# List all Excel files in the directory
  # os.listdir(directory): This function returns a list of all the files and directories in the specified directory
  # list comprehension, which is a concise way to create lists in Python
excel_files = [f for f in os.listdir(directory) if f.endswith('.xlsx')]

# Initialize a list to hold valid DataFrames and their filenames
valid_files = []

# Loop over the Excel files to collect valid DataFrames
for file in excel_files:

    # to read the file it is necessary to join the full directory and file name 
    file_path = os.path.join(directory, file)
    try:
        df = pd.read_excel(file_path, engine='openpyxl')
        
        # Strip any whitespace from column names
        df.columns = df.columns.str.strip()

        # Check for required columns
          # The condition ensures that only DataFrames with the required columns are considered valid.
        if 'time (s)' in df.columns and 'number in man made hole' in df.columns:
            # append tuple which includes the filename and the corresponding DataFrame
            valid_files.append((file, df))
    
    except Exception as e:
        continue 

# Number of valid DataFrames 
num_valid_files = len(valid_files)

if num_valid_files > 0:
    # Create subplots based on the number of valid files
      # squeeze=False ensures that the axes array always has a consistent shape (good cause different numbers of files in different conditions but want same size)
    fig, axes = plt.subplots(nrows=1, ncols=num_valid_files, figsize=(5*num_valid_files, 5), squeeze=False)

    # Plot each valid DataFrame
    for i, (file, df) in enumerate(valid_files):
        axes[0, i].plot(df['time (s)'], df['number in man made hole'])
        axes[0, i].set_xlabel('Time (s)')
        axes[0, i].set_ylabel('Number of Larvae in Side Hole')
        axes[0, i].set_title(file)
        axes[0, i].set_ylim(0, 10)  # Set y-axis limit

    # Add main title
    fig.suptitle('Food Side Hole N10', fontsize=20, fontweight='bold')

    # Adjust layout
    plt.tight_layout()  # Adjust layout to make room for the main title

    # Show the plot
    plt.show()
else:
    print("No valid Excel files found with the required columns 'time (s)' and 'number in man made hole'.")
```
